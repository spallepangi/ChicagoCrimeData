```python
sqlCreateTable = """
CREATE TABLE IF NOT EXISTS spnf4.location_coordinate(
lc_id INT NOT NULL PRIMARY KEY,
x_coordinate INT,
y_coordinate INT,
latitude FLOAT,
longitude FLOAT,
location VARCHAR(50)
);"""
cursor.execute(sqlCreateTable)
connection.commit()
```

```python
sqlCreateTable = """
CREATE TABLE IF NOT EXISTS spnf4.crime_location(
cl_id INT NOT NULL PRIMARY KEY,
la_id INT,
lb_id INT,
lc_id INT,
ld_id INT,
FOREIGN KEY (la_id)
    REFERENCES location_area(la_id),
FOREIGN KEY (lb_id)
    REFERENCES location_block(lb_id),
FOREIGN KEY (lc_id)
    REFERENCES location_coordinate(lc_id),
FOREIGN KEY (ld_id)
    REFERENCES location_description(ld_id)
);"""
cursor.execute(sqlCreateTable)
connection.commit()
```

```python
sqlCreateTable = """
CREATE TABLE IF NOT EXISTS spnf4.crime_record(
id INT NOT NULL PRIMARY KEY,
iucr VARCHAR(50),
fbi_code VARCHAR(50),
cl_id INT,
unnamed_id INT,
case_number VARCHAR(50),
arrest VARCHAR(10),
domestic VARCHAR(10),
crime_date TIMESTAMP,
year INT,
updated_on TIMESTAMP,
CONSTRAINT crime_code_pkey
FOREIGN KEY (iucr,fbi_code)
    REFERENCES crime_code(iucr,fbi_code),
FOREIGN KEY (cl_id)
    REFERENCES crime_location(cl_id)
);"""
cursor.execute(sqlCreateTable)
connection.commit()
```

## 2.5 Developed five distinct and useful queries that could be potentially used for policing planning, policy making, citizen awareness, etc.

* Queries provides some analytic values using GROUP BY/HAVING, Nested Queries, Aggregation Operators, etc.
* Each query has documentation to explain what this query is looking for and why it is meaningful for what purposes.

```python
%%sql
SELECT la.community_area ,COUNT(la.community_area) AS FREQUENCY
FROM crime_record cr 
JOIN crime_location cl ON cl.cl_id = cr.cl_id
JOIN location_area la ON cl.la_id = la.la_id
JOIN crime_code cc ON cc.iucr = cr.iucr AND cc.fbi_code = cr.fbi_code
GROUP BY la.community_area
ORDER BY COUNT(la.community_area) DESC
LIMIT 10
```

```python
%%sql
SELECT lb.block,cc.description, COUNT(*) FROM crime_code cc
JOIN crime_record cr ON cr.iucr = cc.iucr AND cr.fbi_code = cc.fbi_code
JOIN crime_location cl ON cl.cl_id = cr.cl_id
JOIN location_block lb ON lb.lb_id = cl.lb_id
GROUP BY lb.block,cc.description
ORDER BY 3 DESC
```

```python
%%sql

SELECT tbl.community_area,tbl.description, tbl.count, rank FROM 
(
SELECT la.community_area,cc.description, COUNT(*) AS count, RANK() OVER(PARTITION by la.community_area ORDER BY COUNT(*) DESC) FROM crime_code cc
JOIN crime_record cr ON cr.iucr = cc.iucr AND cr.fbi_code = cc.fbi_code
JOIN crime_location cl ON cl.cl_id = cr.cl_id
JOIN location_area la ON la.la_id = cl.lb_id
GROUP BY la.community_area,cc.description
ORDER BY 3 DESC
) AS tbl
WHERE rank<=2
ORDER BY 1,4
```

```python
%%sql
SELECT primary_type,case_number,description
FROM crime_code JOIN crime_record USING(iucr)
WHERE primary_type='OFFENSE INVOLVING CHILDREN' AND description LIKE '%CHILD ABDUCTION%'
```

```python
%%sql
SELECT DISTINCT primary_type,location_description,iucr
FROM crime_code JOIN crime_record USING(iucr) JOIN
crime_location USING(cl_id) JOIN
location_description USING(ld_id)
WHERE location_description='RESIDENCE';
```
